```python
import pandas as pd
excel = pd.read_excel(r'D:\MSci_Project\3_Processed Data\2023-02-15 Bead phantom reconstructions\line-profiles.xlsx')
```

```python
f_a = excel['f_a'].dropna().values[1:]
f_c = excel['f_c'].dropna().values[1:]
f_f = excel['f_f'].dropna().values[1:]
import numpy as np
f_a/= np.max(f_a)
f_c/= np.max(f_c)
f_f/= np.max(f_f)


```

```python
import matplotlib.pyplot as plt
arrays = [f_a, f_c, f_f]
names = ['a', 'c', 'f']
for i in range(3):
    array = arrays[i]
    x_range = (np.arange(0,len(array))-np.argmax(array)) * 6.45 #microns
    plt.plot(x_range,  array, label=names[i])
plt.legend()

plt.xlim(-50, 50)
plt.grid()
plt.xlabel('Coordinate (microns)')
```
