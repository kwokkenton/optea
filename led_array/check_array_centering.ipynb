```python
import numpy as np
from skimage import io

stack = io.imread(r'D:\MSci_Project\1_Data\2023-03-10\2023-03-10 square raster.tiff')
stack_average = np.average(stack, axis=1)
stack_average = np.average(stack_average, axis=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(stack_average)
```

```python
brightness = np.zeros(100)

for i in range(100):
    brightness[i] = stack_average[i]
    
brightness = np.reshape(brightness, (10,10))
```

```python
plt.imshow(brightness, cmap='inferno')
plt.plot(4.5, 4.5, 'o')
plt.axhline(y=4.5)
plt.axvline(x=4.5)
plt.colorbar()
```
