```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Xs = (np.arange(0,15)-7)*4*np.ones((15, 15))
Ys = Xs.T
```

```python
rs = np.sqrt(Xs**2 + Ys**2)
H= 400 # mm
NAs = rs/ np.sqrt(rs**2 + H**2)
```

```python
f_num = 20
thresholded = (NAs<0.5/f_num)
plt.imshow(thresholded)
plt.colorbar()
```
