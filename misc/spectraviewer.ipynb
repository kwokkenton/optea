```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('FluoSpheres yellow-green fluorescent microspheres.csv')
```

```python
data
```

```python
plt.rcParams.update({'font.size': 18})
```

```python
plt.figure(figsize=(12, 5))
plt.plot(data.Wavelength.values, data.Excitation.values, '--', alpha=0.5, label='fluorophore excitation', color='blue')
plt.fill_between(data.Wavelength.values, data.Emission.values, alpha=0.5, label='fluorophore emission', color='green')
plt.axvspan(457, 487, alpha=0.3, color='blue', label='excitation filter')
plt.axvspan(542, 582, alpha=0.3, color='yellow', label='emission filter')
plt.xlim(375, 650)
plt.legend()
plt.xlabel('wavelength(nm)')
plt.ylabel('intensity (%)')
plt.tight_layout()
plt.savefig('spectrum.png', transparent=True, dpi=300)
plt.show()
```
