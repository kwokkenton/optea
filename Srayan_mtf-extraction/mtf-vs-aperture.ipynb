```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
path = '../../data_store/2023-01-17 mtf_values/'

mtf_a = pd.read_csv(path+"f_a.csv")
mtf_b = pd.read_csv(path+"f_b.csv")
mtf_c = pd.read_csv(path+"f_c.csv")
mtf_d = pd.read_csv(path+"f_d.csv")
mtf_e = pd.read_csv(path+"f_e.csv")
```

```python
plt.plot(mtf_a['X'], mtf_a['Y'], label='f_a')
plt.plot(mtf_b['X'], mtf_b['Y'], label='f_b')
plt.plot(mtf_c['X'], mtf_c['Y'], label='f_c')
plt.plot(mtf_d['X'], mtf_d['Y'], label='f_d')
plt.plot(mtf_e['X'], mtf_e['Y'], label='f_e')
plt.title('MTF vs aperture')
plt.xlabel('cycles/pixel')
plt.ylabel('mtf')
plt.legend()
plt.show()
```
